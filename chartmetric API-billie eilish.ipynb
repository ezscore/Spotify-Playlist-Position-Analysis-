```python
#get artist tracks from spotify

def get_artist_tracks(api_token, cm_artist_id):
    response = requests.get(url='https://api.chartmetric.com/api/artist/{}/tracks'.format(cm_artist_id),
                       headers={'Authorization' : 'Bearer {}'.format(api_token)})
    tracks_info = response.json()
    track_list = tracks_info['obj']
    return track_list
```

```python

def parse_track_data(track_list):
    song_list = []
    for song in track_list:
        if song['release_dates'] == (type(list)) and (len(song['release_dates']) > 1):
            
            song_tuple = (song['cm_artist'][0], song['cm_track'], song['name'], song['artist_names'][0], 
                          song['release_dates'][0][0])
        elif song['release_dates'] == (type(list)) and (len(song['release_dates']) == 1):
            
            song_tuple = (song['cm_artist'][0], song['cm_track'], song['name'], song['artist_names'][0], 
                          song['release_dates'][0])
        else:
            song_tuple = (song['cm_artist'][0], song['cm_track'], song['name'], song['artist_names'][0], 
                          song['release_dates'])
        song_list.append(song_tuple)
    return song_list
    

#cm_artist,cm_track, track_name, spotify_track_ids, release_dates, danceability, energy,speechiness, acoustiness, instrumentalness, liveness, valence, tempo, loudness
```

```python
headers={'Authorization' : 'Bearer {}'.format(api_token)}

url_params = {'since': '2016-01-08',
                         'limit': 10,
             }

def get_playlist_data(headers, track_id_list, status):
    playlist_list = []
    for track_id in track_id_list:
        
        response = requests.get(url='https://api.chartmetric.com/api/track/{}/spotify/{}/playlists'.format(track_id, status),
                            headers=headers, 
                                params=url_params)
        json_file = response.json()
        data = json_file['obj']
        playlist_list.extend(data)
    return playlist_list
    
```

```python
past_results = get_playlist_data(headers, track_id_list, 'past')
```

```python
past_results
```

```python
def parse_playlist_activity(results):

    playlist_list = []

    for playlist in results:
        
        remove_list = playlist['playlist'].get('removed_at', 'N/A')
        playlist_tuple = (playlist['playlist']['track_id'],playlist['track']['name'], playlist['playlist']['name'], playlist['playlist']['official'], 
                          playlist['playlist']['followers'], playlist['playlist']['added_at'], remove_list, playlist['playlist']['position'])
        playlist_list.append(playlist_tuple)

    return playlist_list
    
```

```python
#create list of tuples for past playlist positions

past_playlist_results = parse_playlist_activity(past_results)
past_playlist_results
```

```python
len(past_playlist_results)
```

```python
#retrive current playlist results
current_results = get_playlist_data(headers, track_id_list, 'current')
```

```python
#create list of tuples for current playlist positions

current_playlist_results = parse_playlist_activity(current_results)
```

```python
#extend past playlist with current playlist
past_playlist_results.extend(current_playlist_results)
```
