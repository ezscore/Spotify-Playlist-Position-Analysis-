```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('billie_playlist_df.csv', index_col=0)
```

```python
df
```

```python
ocean_eyes_dict = {'ocean eyes':'Ocean Eyes'}
df.track_title = df.track_title.replace(ocean_eyes_dict)
```

```python
df.track_title.unique()
```

```python
df.groupby('track_title')['playlist_position'].mean()
```

```python
df
```

```python
official_playlist_df = df.loc[df['Official'] == True]
```

```python

official_playlist_df.groupby('track_title')['playlist_position'].mean()
```

```python
official_playlist_df.add_date = official_playlist_df.add_date.apply(lambda x: x[0:10])
```

```python
official_playlist_df
```

```python
fig = plt.figure(figsize=(15,8))


official_playlist_df.boxplot(by ='track_title', column =['playlist_position'], figsize=(15,8))
```

```python
fig = plt.figure(figsize=(15,8))

sns.boxplot(data=official_playlist_df, x='track_title', y='playlist_position',  width=.2).set_title('Billie Eilish Playlist Positions')
```
